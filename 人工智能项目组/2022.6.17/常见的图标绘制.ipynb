```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python

np.random.seed(0)
x = np.arange(5)
y = np.random.randint(-5,5,5)
```

```python
# 条形图
fig,axes = plt.subplots(ncols=2)
v_bars = axes[0].bar(x,y,color='r')
h_bars = axes[1].barh(x,y,color='b')
```

```python
plt.bar(x,y)
```

```python
mean_values = [1,2,3]
variance = [0.2,0.4,0.5]
bar_label = ['bar1','bar2','bar3']
x_pos = list(range(len(bar_label)))
plt.bar(x_pos,mean_values,yerr=variance,alpha=0.4)
plt.xticks(x_pos,bar_label)
```

```python
data = range(200,225,5)
bar_label = list('abcde')
fig = plt.figure(figsize=(10,8))
y_pos = np.arange(len(bar_label))
plt.yticks(y_pos,bar_label)
plt.barh(y_pos,data,alpha=0.4,color='b')
plt.vlines(min(data),1,len(data)+0.7,linestyle='--')
```
