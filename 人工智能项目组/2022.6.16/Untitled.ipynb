```python
x = np.linspace(-10,10)
y = np.sin(x)
plt.plot(x,y,linewidth=3.0)
```

```python
plt.plot(x,y,color='b',linestyle=':',marker='o',markfacecolor='r')
```

```python
line = plt.plot(x,y)
plt.setp(line,color='r',linewidth=2.0,alpha=0.4)
```
