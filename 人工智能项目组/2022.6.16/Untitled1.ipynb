```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10,0.5)
y = np.sin(x)
```

```python
plt.subplot(211)
plt.plot(x,y,color='r')
plt.subplot(212)
plt.plot(x,y,color='b')
```

```python
plt.subplot(121)
plt.plot(x,y,color='r')
plt.subplot(122)
plt.plot(x,y,color='g')
```

```python
plt.subplot(321)
plt.plot(x,y,color='r')
plt.subplot(324)
plt.plot(x,y,color='r')
```

```python
plt.plot(x,y,color='b',linestyle=':',marker='o',markerfacecolor='r',markersize=10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('hxx')
#plt.text(0,0,'happy')
plt.annotate(text='happy',xy=(0.2,0),xytext=(1.5,0),arrowprops=dict(facecolor='b',shrink=0.05,headlength=20,headwidth=10))
```

```python
x = range(10)
y = range(10)
plt.grid(True)
plt.plot(x,y)
```

```python
plt.grid(True)
fig = plt.gca()
plt.plot(x,y)
plt.grid(False)
fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
```

```python
import math
x = np.random.normal(loc=0.0,scale=1.0,size=300)
width = 0.5
bins = np.arange(math.floor(x.min())-width,math.ceil(x.max())+width,width)
ax = plt.subplot(111)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tick_params(bottom=False,top=False,left=False,right=False)

plt.grid(True)
plt.hist(x,alpha=0.5,bins=bins)
```

```python
x = range(10)
y = range(10)
label = ['ok' for i in range(10)]
fig.ax = plt.subplots()
plt.plot(x,y)
plt.title('hxx kc')
ax.set_xticklabels(labels=label,rotation=45)
```
