### 信息增益


目标：构建决策树模型<br>
天气情况：1晴天、雨天、阴天<br>
气温情况：高、适中、凉爽<br>
湿度情况：高、普通<br>
 风情况：有风、没风<br>
 

在历史数据中有8天出去玩，有6天不出去玩，此时还没有划分数据集，熵值：<br>
-8/14 * log2 * 8/14 - 6/14 * log2 * 6/14 = 0.98<br>
当天气情况为晴天的时候：共计4条（3天不出去、1天出去）<br>
-3/4 * log2 * 3/4 - 1/4 * log2 * 1/4 = 0.8112<br>
当天气情况为雨天的时候：共计6条（2天不出去、4天出去）<br>
-2/6 * log2 * 2/6 - 4/6 * log2 * 4/6 = 0.9239<br>
当天气情况为阴天的时候：共计4条（全部为出去玩）<br>
熵值为：0<br>
天气情况：为晴天、雨天、阴天他们的概率为4/14、6/14、4/14，计算天气情况为节点的熵值：<br>
4/14 * 0.8112 + 6/14 * 0.9239 + 4/14 * 0 = 0.6816<br>

当以天气情况为根节点时，信息增益如下：<br>
0.98 - 0.6816 = 0.2984<br>
